```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
# load data
Data = pd.read_csv('train.csv')
```

```python
print(Data.shape)
Data.head()
```

```python
train_outputs = Data['SalePrice']
Data = Data.drop(['SalePrice'], axis=1)
```

```python
# For some features, 'NA' is an acceptable value, so change the pd.NA 
# for those to 'NA' where necessary
# list of columns that are supposed to have NA
acceptable_na = ['Alley', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                 'BsmtFinType2', 'FireplaceQu', 'GarageType', 'GarageFinish', 
                 'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']

for feature in acceptable_na:
    Data[feature].fillna('NA', inplace=True)
```

```python
remove_columns = ['Id', 'Utilities', 'Condition1', 'Condition2', 'PoolArea', 'PoolQC', 'MiscFeature']
Data = Data.drop(remove_columns, axis=1)
```

```python
for col in Data.columns:
    print(col)
```

## basement best features selection

```python
bsmt_cols = [col for col in Data if col.startswith('Bsmt')]
bsmt_cols.append('TotalBsmtSF')
bsmt_data = Data[bsmt_cols]
bsmt_data.head()
```

```python
ords = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']


cats1 = ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex']
cats2 = ['NA', 'No', 'Mn', 'Av', 'Gd']
cats3 = ['NA', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ']
encoder1 = preprocessing.OrdinalEncoder(categories=[cats1, cats1, cats2, cats3, cats3])

bsmt_data[ords] = encoder1.fit_transform(bsmt_data[ords])
```

```python
bsmt_data.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
X = bsmt_data.to_numpy()
y = train_outputs.to_numpy()
reg = DecisionTreeRegressor()
#cross_val_score(reg, X, y, cv=10)
```

```python
reg.fit(X,y)
```

```python
reg.feature_importances_
```

Based off the above the most important bsmt features are BsmtQual and TotalBsmtSF.
