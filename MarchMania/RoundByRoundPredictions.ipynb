```python
import pandas as pd
from IPython.display import display

# ---------------------------------------------------------------------------------------------------------------

def see_prediction(name1, name2, year=2021):
    files = ['chalk', 'simple_knn_10_neighbors']
    all_preds = pd.DataFrame()
    
    for f in files:
        preds = pd.read_csv(f'submissions/{year}/{f}.csv')
    
        matchups = pd.read_csv('data/TournamentMatchupsWNameDF_2021.csv')
        matchups['ID'] = matchups.apply(lambda x: f'{int(x.Season)}_{(x.ATeamID)}_{x.BTeamID}', 1)
        matchups.drop(columns=['Season', 'ATeamID', 'BTeamID'], inplace=True);

        preds = matchups.merge(preds, on=['ID'])
        preds['Winner'] = preds.apply(lambda row: row.ATeamName if (row.Pred > 0.5) else row.BTeamName, axis=1)
        pred  = preds.loc[((preds['ATeamName']==name1) & (preds['BTeamName']==name2))]
        
        all_preds = pd.concat([all_preds, pred], ignore_index=True)
    
    index = {i: files[i] for i in range(len(files))}
    display(all_preds.rename(index=index))
```
